```python
# Import the libraries you'll need here.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import scipy.stats as stats
from IPython.display import Image
```

# Load the cars csv file into a pandas dataframe.


```python
# Load the csv file into a pandas dataframe.

df = pd.read_csv("data/cars.csv")
df
```

# Data clean up part 1.

1. Print the number of duplicate rows we have in our dataframe.

2. Modify our df to have all duplicate rows removed. 

3. Do a sanity check to make sure all duplicates have been removed by printing the total number of duplicate rows again.

```python
# 1. Print the number of duplicate rows we have in our dataframe.

print(df.duplicated().sum())

df.duplicated()
```

```python
#  2. Modify our df to have all duplicate rows removed. 
df1 = df[df.duplicated() == False]
df1
```

```python
# 3. Do a sanity check to make sure all duplicates have been removed by printing the total number of duplicate rows again.

print(df1.duplicated().sum())
```

# Data clean up part 2.
* Which column has the most null values and how many null values does it have?
* Print how long our dataframe is.
* Remove any row that has a null value in it. 
* Do a sanity check and pring how long our dataframe is now that we have removed our null values.

```python
# * Which column has the most null values and how many null values does it have?
df1.isnull().sum()
```

```python
# * Print how long (aka how many rows) our dataframe is.

print(df1.shape)
```

```python
# * Remove any row that has a null value in it. 

df2 = df1.dropna()
```

```python
# * Do a sanity check and pring how long our dataframe is now that we have removed our null values.

df2.isnull().sum()
```

### Make a timeline line chart in which the x-axis is the year, and the y-axis is the average MSRP.
* What's noticeable about it and what do you think the error is...


```python
# Make a timeline line chart in which the x-axis is the year, and the y-axis is the average MSRP.

mrsp = df2.groupby('Year')['MSRP'].mean()
mrsp.plot(kind='line',x ='Year', y ='MSRP' , figsize=(10,10) )
```

# It seems as though in the years before 2000, they were counting in tens.
Make a new column that is called `adjusted_price`, that contains all prices, however, for 
every year before 2000 make it 10x the original MSRP.  
_Hint use `np.where()`_.

```python
# Make a new column that is called `adjusted_price`, that contains all prices, however, for every year before 2000 make it 10x the original MSRP.

c1 = df2['Year']< 2000

df2['adjusted_price']=np.where(c1,df2['MSRP']*10,df2['MSRP'])
df2
```

# Replot the new adjusted price.  
* Make the y-axis start at 0 and go up to 100,000

```python
# Replot the new adjusted price and make the y-axis start at 0 and go up to 100,000
ax=sns.lineplot(x="Year", y="adjusted_price", data=df2)
ax.set(ylim=(0,100000))
```

# Remove any statisical outliers from Engine HP and highway MPG and plot the result.
* Statisical outliers meaning values that are further than 3 standard deviations from the mean 
* Create a new columns for z-scores for each 'Engine HP' and 'highway MPG' named 'Engine HP_zscore' and 'highway MPG_zscore'
* Calculate the z-scores for each of our target columns.
* Make sure you set the z-scores to be their absolute values. 
* Create condition masks for when either of those absolute values are greater than 3.

* Create a new dataframe that is a copy of our dataframe using df.copy()

* Using our new dataframe
    * Filter out all Engine HP Z-Scores that are greater than 3
    * Filter out all Highway MPG z-scores that are greater than 3.
    * Make the same scatterplot plotting the relationship of Engine HP and Highway MPG as before but with the this dataframe.

```python
# Remove any statisical outliers from Engine HP and highway MPG

new_col = ['Engine HP', 'highway MPG']
z_score_cols = []

for col in new_col:
    new_col_name = col + '_zscore'
    df2[new_col_name] = stats.stats.zscore(df2[col])
    df2[new_col_name] = abs(df2[new_col_name])
    z_score_cols.append(new_col_name)
    
    
dfc = df2.copy()
eng_cond = dfc['Engine HP_zscore'] <= 3
hig_cond = dfc['highway MPG_zscore'] <=3

new_df = dfc[eng_cond&hig_cond]
sns.scatterplot(data=new_df,x='Engine HP',y='highway MPG')
```
