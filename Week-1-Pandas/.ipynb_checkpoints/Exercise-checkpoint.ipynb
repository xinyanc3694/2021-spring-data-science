## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
less = df["price"] < 100

df[less].id.count()
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

bk = df['neighbourhood_group'] == "Brooklyn"

df_bk = df[bk]

df_bk.id.count()
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

bk_less = df[less & bk]
bk_less.id.count()

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
name_list = ["Michael","David","John","Daniel"]

condition = df.host_name.isin(name_list)

dfn = df[condition]
dfn
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

c1 = df['neighbourhood'] == 'Williamsburg'

df['adjusted_price'] = np.where(c1, df["price"]+100, df["price"])

df

```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  
dfp = (df.room_type.value_counts()/ df.room_type.value_counts().sum())*100
dfp
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
df.groupby('neighbourhood_group')['id'].count()

```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

df.groupby('neighbourhood_group')['price'].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
df.groupby('neighbourhood_group')['price'].agg(['min', 'max'])

```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
df.groupby('room_type')['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.


gby = ['neighbourhood_group','room_type']
df.groupby(gby)['price'].agg(['count', 'min', 'max', 'mean', 'median', 'std'])
```

# Join and file saving.
1. Load the `prices.csv` and the `n_listings.csv`


2. Do join that keeps all the records for each table.
    * Neighbourhood groups should include ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island',
       'LongIsland']
       
       
3. Save your joined csv as `joined.csv`


4. Load your saved table and see if it looks the same or different that the DataFrame you used to create it. 

```python
# 1. Load the `prices.csv` and the `n_listings.csv`

dfprices = pd.read_csv('data/prices.csv',sep=",")
dflistings = pd.read_csv('data/n_listings.csv',sep=";")
```

```python
# 2. Do join that keeps all the records for each table.
dfjoined = pd.merge(dfprices, dflistings, on='neighbourhood_group', how='outer')

dfjoined
```

```python
#3. Save your joined csv as joined.csv
dfjoined.to_csv('joined.csv',index=False)
```

```python
#4. Load your saved table and see if it looks the same or different that the DataFrame you used to create it.
dfx = pd.read_csv("joined.csv")

dfx
```

1. Who was won Album of the Year in 2016?

```python
#  1. Who was won Album of the Year in 2016?

dfg = pd.read_csv("Data/grammys.csv")

year = dfg['year'] == 2016
cate = dfg.category == "Album of the Year"
win = dfg.winner == True
new_dfg = dfg[year & cate & win]
new_dfg
```

```python
# 2. Who won Best Rap Album in 2009?

year = dfg['year'] == 2009
cate = dfg.category == "Best Rap Album"
win = dfg.winner == True
new_dfg_2 = dfg[year & cate & win]
new_dfg_2
```
